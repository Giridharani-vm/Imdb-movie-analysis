6... top rating movies at each genre

```python
import pandas as pd
imdb_movies = pd.concat([ad,bio,fan,rom,sci])
imdb_movies.shape
imdb_movies.to_csv("imdb_movies.csv",index=False)
```

```python
imdb_movies = imdb_movies[imdb_movies["Rating"] > 0.0]
imdb_movies['Rating'] = imdb_movies['Rating'].astype(float)
sorted_movies = imdb_movies.sort_values(by=['Genre', 'Rating'], ascending=[True, False])
top_rating = sorted_movies.drop_duplicates(subset='Genre', keep='first')[['Genre', 'Title', 'Rating']]
display(top_rating)
```

7. highest votings in pie chart

```python
d = [ad,bio,fan,rom,sci]
for i in d:
    print(round(i["Votes"].max()))
```

8...shortest and longest movies table/card

```python
imdb_movies['Duration'] = imdb_movies['Duration'].astype(float).round(2)
imdb_movies = imdb_movies[imdb_movies['Duration'] > 0.0]
longest_movies = imdb_movies.sort_values(by=['Genre', 'Duration'], ascending=[True, False]).drop_duplicates('Genre')[['Genre', 'Title', 'Duration']]
shortest_movies = imdb_movies.sort_values(by=['Genre', 'Duration'], ascending=[True, True]).drop_duplicates('Genre')[['Genre', 'Title', 'Duration']]
longest_movies.reset_index(drop=True, inplace=True)
shortest_movies.reset_index(drop=True, inplace=True)
display(longest_movies)
display(shortest_movies)

```

```python
long_all = imdb_movies["Duration"].max()
short_all = imdb_movies["Duration"].min()
print(long_all)
print(short_all)
```
